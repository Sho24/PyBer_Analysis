```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis =["Jan","Feb","Mar","Apr","May","June","July","Aug","Sep","Oct","Nov","Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot with ax.plt()
fig,ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()  (Object-Oriented)
ax = plt.axes()
ax.plot(x_axis, y_axis,marker="*",color="green",linewidth=2,label="Boston")
ax.legend("Boston")
ax.set_title("Pyber Fare by Months")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_ylim(0,45)
ax.legend()
ax.grid()
```

```python

#Create the plot and add a lable for the legend
plt.plot(x_axis,y_axis,marker="*",color="blue",linewidth=2,label="Boston")

#Add Lable to X-axis and Y-axis 
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Set the x-axislimits from Jan to June
#plt.xlim(0,5)

#Set the y-axis limits between 0 - 45
plt.ylim(0,45)

#Create title
plt.title("PyBer Fare by Month")

#Add the legend
plt.legend()
```

```python
# Create the plot
plt.bar(x_axis, y_axis,color="green",label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot (Horizontal Bar Chart)
plt.barh(x_axis, y_axis)
```

```python
plt.barh(x_axis,y_axis,color="magenta",label="Boston")
plt.legend()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.gca().invert_yaxis()
```

```python
# Create the plot with ax.plt() (Object-Oriented)
fig,ax=plt.subplots()
ax.bar(x_axis,y_axis,color="k",label="Boston")
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis,x_axis,s=[ i * 5 for i in y_axis],marker='o',alpha=0.20,edgecolors='black',linewidth=2,label="Boston")
ax.legend()
ax.set_xlim(0,45)
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
plt.gca().invert_yaxis()
```

```python
plt.pie(y_axis,labels=x_axis)
plt.show()
```

```python

explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```
